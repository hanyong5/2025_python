```python
import csv
import matplotlib.pyplot as plt

# data = csv.reader(open('seouldata3.csv',encoding='utf-8')) #[[],[],[]]

# count = 0
# for row in data:
#     # print(row)
#     count += 1
# print(count)
tcount=0
count = 0

htemp = []

with open('seouldata3.csv',encoding='utf-8') as f:
    data=csv.reader(f)
    next(data)
    for row in data:
        tcount += 1
        if len(row) > 0 and row[-1] != '':
            count += 1
            htemp.append(float(row[-1]))

print(f'데이터 행수 : 총{tcount} /  최고온도{count}')
# print(type(htemp[0]))
# print(htemp[:10])

plt.figure(figsize=(2,4))
# plt.plot(htemp)
plt.boxplot(htemp)
plt.title('htemp')
plt.ylabel('temp')
plt.grid(True,linestyle='--',linewidth=0.5, color='gray',alpha=0.5,axis='y')
plt.show()

```
