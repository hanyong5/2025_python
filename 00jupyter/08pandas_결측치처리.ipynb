```python
titanic_df = pd.read_csv('titanic_train.csv')
titanic_df.head()
```

```python
# titanic_df.info()
titanic_df.isna().sum()
```

```python
# titanic_df["Age"].mean()
titanic_df['Age'] = titanic_df['Age'].fillna(titanic_df["Age"].mean())
# titanic_df.isna().sum()
titanic_df.head()
```

```python
most_cabin = titanic_df['Embarked'].value_counts()
most_cabin
```

```python
titanic_df['Embarked'] = titanic_df['Embarked'].ffill()
titanic_df.isna().sum()
```

```python
titanic_df.drop('Cabin',axis =1,inplace=True)
```
