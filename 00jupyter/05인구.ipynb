```python
import csv
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

path = 'C:/Windows/Fonts/Noto Sans KR/NotoSansKR-VF.ttf'
fontprop = fm.FontProperties(fname=path, size=18)


f = open('2024.csv',encoding='utf-8')
data = csv.reader(f)
next(data)

result1 = []
result2 = []


name1 = input('첫번째 지역입력 : ')
name2 = input('두번째 지역입력 : ')

for row in data:
    if name1 in row[0]:
        for i in row[3:]:
            result1.append(int(i))
    elif name2 in row[0]:
        for i in row[3:]:
            result2.append(int(i))

plt.plot(result1,label=name1,c='r')
plt.plot(result2,label=name2,c='b')
plt.legend()
plt.show()

```
