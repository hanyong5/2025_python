```python



import matplotlib.pyplot as plt

# 기본 list 데이터
data = [
    100, 105, 98, 110, 90, 95, 102, 108, 120, 85,
    150, 130, 80, 77, 140, 123, 160, 100, 98, 97
]

# 정렬
sorted_data = sorted(data)
n = len(sorted_data)

# 중앙값 (Median)
def median(lst):
    n = len(lst)
    mid = n // 2
    if n % 2 == 0:
        return (lst[mid - 1] + lst[mid]) / 2
    else:
        return lst[mid]

# Q1, Q3
Q2 = median(sorted_data)
Q1 = median(sorted_data[:n // 2])
Q3 = median(sorted_data[(n + 1) // 2:])  # 중간 포함 X

IQR = Q3 - Q1
lower_whisker = max(min(data), Q1 - 1.5 * IQR)
upper_whisker = min(max(data), Q3 + 1.5 * IQR)

# 이상치
outliers = [x for x in data if x < lower_whisker or x > upper_whisker]

# 박스플롯 그리기
fig, ax = plt.subplots()
ax.boxplot(data, vert=True, patch_artist=True)

# 텍스트로 수치 표시
ax.text(1.1, Q1, f'Q1: {Q1:.1f}', va='center')
ax.text(1.1, Q2, f'Median: {Q2:.1f}', va='center')
ax.text(1.1, Q3, f'Q3: {Q3:.1f}', va='center')
ax.text(1.1, lower_whisker, f'Min: {lower_whisker:.1f}', va='center')
ax.text(1.1, upper_whisker, f'Max: {upper_whisker:.1f}', va='center')

# 이상치 표시
for outlier in outliers:
    ax.text(1.05, outlier, f'{outlier}', color='red', fontsize=8)

# 그래프 꾸미기
ax.set_title("Boxplot (list only)")
ax.set_ylabel("Values")
plt.tight_layout()
plt.show()
```

```python
import requests
from bs4 import BeautifulSoup

headers = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.86 Safari/537.36'}
data = requests.get('https://www.melon.com/chart/',headers=headers)

soup = BeautifulSoup(data.text, 'html.parser')

top_100 = soup.select('#lst50 > td > div') #tr로 이루어진 list

for tr in top_100 :
    rank = tr.select_one('span.rank')
        
    if rank is not None:
        print(rank.text, end=". ")
        
    title = tr.select_one('div > div.ellipsis.rank01 > span > a')
    
    if title is not None:
        artist = tr.select_one('div.ellipsis.rank02 > a').text
        print(artist,'-', title.text)
```
