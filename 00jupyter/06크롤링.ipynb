```python
from bs4 import BeautifulSoup


html = '''
<html>
  <body>
    <div id="main">
      <h1 class="title">파이썬 웹 스크래핑</h1>
      <h1>test</h1>
      <p class="desc">BeautifulSoup를 이용한 데이터 추출 예제</p>
      <a href="https://example.com" class="link">더 알아보기</a>
      <ul class="items">
        <li data-id="101">첫 번째</li>
        <li data-id="102">두 번째</li>
        <li data-id="103">세 번째</li>
      </ul>
    </div>
  </body>
</html>
'''
soup = BeautifulSoup(html, "html.parser")

# 1.soup.find("tag")
h1_tag = soup.find('li')
print(h1_tag)


# 2. soup.find("tag",class_="desc")
p_tag = soup.find("p",class_="desc")
print(p_tag.text)

# 3. soup.find_all("tag")
# li_tags = soup.find_all("li")
# print(li_tags)
# for li in li_tags:
#     print("-",li.text)

# 4. soup.select("css 선택자")
css_li = soup.select("ul.items  li")
for li in css_li:
    print("o", li.text)

```

```python
import requests
from bs4 import BeautifulSoup

headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/136.0.0.0 Safari/537.36'}
data= requests.get("https://www.melon.com/chart/",headers=headers)
soup = BeautifulSoup(data.text,'html.parser')

# print(soup)

# title_one = soup.select_one("#tb_list table .lst50 .ellipsis.rank01 a")
# print(title_one.text)

top_50 = soup.select('#lst50 > td > div')

for row in top_50:
    rank = row.select_one('span.rank')
    if rank is not None:
        print(rank.text,end=". ")
  
    title = row.select_one('.ellipsis.rank01 > span > a')
    if title is not None:
        print(title.text,end = " - ")

    singer = row.select_one('.ellipsis.rank02 > span > a')
    if singer is not None:
        print(singer.text)




```
