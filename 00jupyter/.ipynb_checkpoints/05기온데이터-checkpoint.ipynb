```python
import csv
import matplotlib.pyplot as plt

# data = csv.reader(open('seouldata3.csv',encoding='utf-8')) #[[],[],[]]

# count = 0
# for row in data:
#     # print(row)
#     count += 1
# print(count)
tcount=0
count = 0

htemp = []

with open('seouldata3.csv',encoding='utf-8') as f:
    data=csv.reader(f)
    next(data)
    for row in data:
        tcount += 1
        if len(row) > 0 and row[-1] != '':
            count += 1
            htemp.append(float(row[-1]))

print(f'데이터 행수 : 총{tcount} /  최고온도{count}')
# print(type(htemp[0]))
# print(htemp[:10])

plt.figure(figsize=(2,4))
# plt.plot(htemp)
plt.boxplot(htemp)
plt.title('htemp')
plt.ylabel('temp')
plt.grid(True,linestyle='--',linewidth=0.5, color='gray',alpha=0.5,axis='y')
plt.show()

```

```python
import csv
import matplotlib.pyplot as plt

april = []
oct =[]

with open('seouldata3.csv',encoding="utf-8") as f:
    data = csv.reader(f)
    next(data)
    for row in data:
        if len(row) > 0 and row[-1] != "":
            if row[0].split('-')[1] == "04":
                april.append(float(row[-1]))
            if row[0].split('-')[1] == "10":
                oct.append(float(row[-1]))

print(april[:10],oct[:10])
plt.boxplot([april,oct])
plt.show()
            
            
```

```python
import csv
import matplotlib.pyplot as plt
import numpy as np

x = [[] for _ in range(12)]
# print(x)

with open('seouldata3.csv',encoding="utf-8") as f:
    data = csv.reader(f)
    next(data)
    for row in data:
        if len(row) > 0 and row[-1] != "":
            x[int(row[0].split('-')[1])-1].append(float(row[-1]))

# print(x)

for y in range(12):
    print(f'{y+1}월, {len(x[y])}')


# 박스플롯 그리기
box = plt.boxplot(x)

# 중앙값 위치 가져오기
for i, line in enumerate(box['medians']):
    # 중앙값 y좌표 구하기
    median_value = line.get_ydata()[0]
    # 텍스트로 중앙값 표시
    plt.text(i + 1, median_value, f'{median_value:.2f}',
             horizontalalignment='center',
             verticalalignment='bottom',
             fontsize=8, color='red')


plt.boxplot(x)
plt.grid
plt.show()
            
```
